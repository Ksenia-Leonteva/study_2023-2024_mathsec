```python
import numpy as np
import math
```

1. Алгоритм, реализующий тест Ферма

```python
def Fermat(n):
    a = np.random.randint(2, (n - 2) + 1)
    r = (a ** (n - 1)) % n
    if r == 1:
        return "Число " + str(n) + ", вероятно, простое"
    else:
        return "Число " + str(n) + " составное"
```

2. Алгоритм вычисления символа Якоби

```python
def Jacobi(a,n):
    g = 1
    s = 1
    while (a != 0) and (a != 1):
        a1 = a
        k = 0
        if a1 % 2 != 0: 
            k = 0
        if a1 % 2 == 0:
            while a1 % 2 == 0:
                a1 = int(a1 / 2)
            while a != (2 ** k) * a1:
                k = k + 1
        if k % 2 == 0:
            s = 1
        else:
            if (n % 8 == 1 % 8) or (n % 8 == -1 % 8):
                s = 1
            elif (n % 8 == 3 % 8) or (n % 8 == -3 % 8):
                s = -1
        if a1 == 1:
            return g * s

        if (n % 4 == 3 % 4) and (a1 % 4 == 3 % 4):
            s = -s
        a = n % a1
        n = a1
        g = g * s

    if a == 0:
        return 0
    else:
        return g  
```

3. Алгоритм Соловэя-Штрассена

```python
def S_SH(n):
    for i in range(2):
        a = np.random.randint(2, (n - 1) + 1)
        if math.gcd(a,n) > 1:
             return "Число " + str(n) + " составное"
        r = (a ** ((n - 1) / 2)) % n
        s = Jacobi(a,n)
        if r != s % n:
            return "Число " + str(n) + " составное"
    return "Число " + str(n) + ", вероятно, простое"
```

4. Алгоритм Миллера-Рабина

```python
def M_R(n):
    r = n - 1
    s = 0
    if r % 2 != 0: 
        s = 0
    if r % 2 == 0:
        while r % 2 == 0:
            r = int(r / 2)
        while n - 1 != (2 ** s) * r:
            s = s + 1  
    a = np.random.randint(2, (n - 2) + 1)
    y = (a ** r) % n
    if (y != 1) and (y != n - 1):
        j = 1
        if (j <= s - 1) and (y != n - 1):
            y = (y ** 2) % n
            if y == 1:
                return "Число " + str(n) + " составное"
            j = j + 1
        if y != n - 1:
            return "Число " + str(n) + " составное"
    return "Число " + str(n) + ", вероятно, простое"
```
